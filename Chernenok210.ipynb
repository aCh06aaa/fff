## Численное решение ОДУ методом Эйлера

Метод Эйлера — один из базовых численных методов для решения обыкновенных дифференциальных уравнений.
Он позволяет получить приближённое решение задачи Коши:

$$ \frac{dy}{dt} = f(t, y), \quad y(t_0) = y_0 $$

Формула метода:

$$ y_{k+1} = y_k + h \cdot f(t_k, y_k) $$

где $h$ — шаг интегрирования.

## Модель хищник-жертва (Вольтерра–Лотка)

$$
\begin{cases}
\frac{dN}{dt} = rN - aNP \\
\frac{dP}{dt} = bNP - mP
\end{cases}
$$

Где:
- N — жертвы
- P — хищники

```python
import numpy as np
import matplotlib.pyplot as plt

# параметры
alpha = 1.0
beta = 0.1
gamma = 0.05
delta = 0.4

# начальные условия
prey0 = 100
pred0 = 10

# время
t0 = 0
t_end = 50
h = 0.01

steps = int((t_end - t0) / h)

t = np.linspace(t0, t_end, steps)
prey = np.zeros(steps)
pred = np.zeros(steps)

prey[0] = prey0
pred[0] = pred0

# метод Эйлера
for i in range(steps - 1):
    dN = alpha * prey[i] - beta * prey[i] * pred[i]
    dP = gamma * prey[i] * pred[i] - delta * pred[i]

    prey[i+1] = prey[i] + h * dN
    pred[i+1] = pred[i] + h * dP

# график
plt.plot(t, prey, label='Жертвы')
plt.plot(t, pred, label='Хищники')
plt.xlabel('Время')
plt.ylabel('Численность')
plt.legend()
plt.grid()
plt.show()
```
