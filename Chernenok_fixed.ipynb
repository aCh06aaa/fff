# Численное решение ОДУ методом Эйлера

## 1. Введение
В данной работе рассматривается численное решение системы дифференциальных уравнений методом Эйлера на примере модели хищник–жертва (Лотки–Вольтерры).

## 2. Математическая модель

Рассматривается система:

$$
\begin{cases}
\frac{dN}{dt} = rN - aNP \\
\frac{dP}{dt} = bNP - mP
\end{cases}
$$

где $N$ — популяция жертв, $P$ — популяция хищников.

## 3. Реализация метода Эйлера
Метод Эйлера аппроксимирует решение по формуле:

$$ y_{n+1} = y_n + h f(t_n, y_n) $$

```python
import numpy as np
import matplotlib.pyplot as plt

# параметры модели
alpha = 1.0
beta = 0.1
gamma = 0.05
delta = 0.4

# начальные условия
prey0 = 100
pred0 = 10

# временные параметры
t0 = 0
t_end = 50
h = 0.01

steps = int((t_end - t0) / h)

t = np.linspace(t0, t_end, steps)
prey = np.zeros(steps)
pred = np.zeros(steps)

prey[0] = prey0
pred[0] = pred0

# метод Эйлера
for i in range(steps - 1):
    prey[i+1] = prey[i] + h * (alpha * prey[i] - beta * prey[i] * pred[i])
    pred[i+1] = pred[i] + h * (gamma * prey[i] * pred[i] - delta * pred[i])
```

## 4. Визуализация результатов

```python
plt.figure()
plt.plot(t, prey, label="Жертвы")
plt.plot(t, pred, label="Хищники")
plt.xlabel("Время")
plt.ylabel("Популяция")
plt.legend()
plt.grid()
plt.show()
```

## 5. Заключение
Метод Эйлера позволяет получить качественное приближение решения системы. Однако для повышения точности могут использоваться более сложные методы (например, Рунге–Кутты).
